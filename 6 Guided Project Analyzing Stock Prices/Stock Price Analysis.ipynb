## Computing Aggregates

* Calculate a dictionary where the keys are the stock symbols and the values are the average closing price of that stock.
* Calculate the stock symbol with the minimum average closing price.
* Calculate the stock symbol with the maximum average closing price.

```python
d = {}
for stock in stock_prices.keys():
    d[stock] = [stock_prices[stock]['close'].mean()]
    d[stock].append(stock_prices[stock]['close'].min())
    d[stock].append(stock_prices[stock]['close'].max())
```

### Sorting the dictionary by the AVG., MIN and MAX closing price

```python
dict(sorted(d.items(), key=lambda item: item[1][0], reverse=True))
```

```python
dict(sorted(d.items(), key=lambda item: item[1][1], reverse=True))
```

```python
dict(sorted(d.items(), key=lambda item: item[1][2], reverse=True))
```

* Highest MIN Closing Prices:

* Highest MAX Closing Prices:
    * AMZN and BIDU

### Calculate a dictionary where the keys are the days and the values are lists of pairs (volume, stock_symbol) of all trades that occurred on that day.


```python
trades_by_day = {}

for stock_sym in stock_prices:
    for index, row in stock_prices[stock_sym].iterrows():
        day = row["date"]
        volume = row["volume"]
        pair = (volume, stock_sym)
        if day not in trades_by_day:
            trades_by_day[day] = []
        trades_by_day[day].append(pair)
```

```python
trades_by_day['2007-01-03']
```

### Finding The Most Traded Stock Each Day

Calculate a dictionary there the keys are the days and the value of each day is a pair (volume, stock_symbol) with the most traded stock symbol on that day.

```python
most_traded_by_day = {}

for day in trades_by_day:
    trades_by_day[day].sort()
    most_traded_by_day[day] = trades_by_day[day][-1]
```

### Searching For High Volume Days

```python
daily_volumes = []

for day in trades_by_day:
    day_volume = sum([volume for volume, _ in trades_by_day[day]])
    daily_volumes.append((day_volume, day))

daily_volumes.sort()

daily_volumes[-10:]
```

### Finding Profitable Stocks

```python
percentages = []

for stock_sym in stock_prices:
    prices = stock_prices[stock_sym]
    initial = prices.loc[0, "close"]
    final = prices.loc[prices.shape[0] - 1, "close"]
    percentage = 100 * (final - initial) / initial
    percentages.append((percentage, stock_sym))

percentages.sort()

percentages[-10:]
```

The most profitable stock to buy in 2007 would have been ADMP, which appreciated from around 7 cents to its current price of 4.43.
